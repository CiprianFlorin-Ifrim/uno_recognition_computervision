```python
from tkinter import *
from PIL import Image, ImageTk
import cv2
import numpy as np

image_name = './uno_images/b0.jpg'

image = cv2.imread(image_name)

#Rearrang the color channel
b,g,r = cv2.split(image)
img = cv2.merge((r,g,b))

# A root window for displaying objects
root = Tk()  

# Convert the Image object into a TkPhoto object
im = Image.fromarray(img)
imgtk = ImageTk.PhotoImage(image=im) 

# Put it in the display window
Label(root, image=imgtk).pack() 

root.mainloop() # Start the GUI




```

```python
from tkinter import *
from PIL import ImageTk, Image
import cv2

# Create object 
root = Tk()
 
# Adjust size 
root.geometry("1800x480")

# Window Title
root.title('COMPUTER VISION - UNO IDENTIFICATION')

label1 = Label(root)
label1.place(x = 0,y = 0)

# Capture from camera
cap = cv2.VideoCapture(0)

# function for video streaming
def video_stream():
    _, frame = cap.read()
    cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
    img = Image.fromarray(cv2image)
    imgtk = ImageTk.PhotoImage(image=img)
    label1.imgtk = imgtk
    label1.configure(image=imgtk)
    label1.after(1, video_stream) 


video_stream()
root.mainloop()
```

```python
# import the necessary packages
from tkinter import *
from PIL import Image
from PIL import ImageTk
from tkinter.filedialog import askdirectory
from tkinter.filedialog import askopenfilename
import cv2
from IPython.core.display import display, HTML                                    
display(HTML("<style>.container { width:100% !important; }</style>"))                                                #change width of Jupyer Notebook to use the whole window resolution available

def select_image():
    # grab a reference to the image panels
    global panel1, panel2, panel3, panel4, panel5, panel6, panel7, panel8, panel9
    # open a file chooser dialog and allow the user to select an input
    # image
    path = askopenfilename()
    # ensure a file path was selected
    if len(path) > 0:
        # load the image from disk, convert it to grayscale, and detect
        # edges in it
        image = cv2.imread(path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        edged = cv2.Canny(gray, 50, 100)
        # OpenCV represents images in BGR order; however PIL represents
        # images in RGB order, so we need to swap the channels
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # convert the images to PIL format...
        image = Image.fromarray(image)
        edged = Image.fromarray(edged)
        # ...and then to ImageTk format
        image = ImageTk.PhotoImage(image)
        edged = ImageTk.PhotoImage(edged)
        #if the panels are None, initialize them
        if panel1 is None or panel2 is None or panel3 is None or panel4 is None or panel5 is None or panel6 is None or panel7 is None or panel8 is None or panel9:
            # the first panel will store our original image
            panel1 = Label(image=image)
            panel1.image = image
            panel1.pack(side="left", padx=10, pady=10)
            # while the second panel will store the edge map
            panel2 = Label(image=edged)
            panel2.image = edged
            panel2.pack(side="right", padx=10, pady=10)
            
            panel3 = Label(image=image)
            panel3.image = image
            panel3.pack(side="right", padx=10, pady=10)
            
            panel4 = Label(image=edged)
            panel4.image = edged
            panel4.pack(side="right", padx=10, pady=10)
            
            panel5 = Label(image=edged)
            panel5.image = edged
            panel5.pack(side="right", padx=10, pady=10)
        # otherwise, update the image panels
        else:
            # update the pannels
            panel1.configure(image=image)
            panel2.configure(image=edged)
            panel3.configure(image=image)
            panel4.configure(image=edged)
            panel5.configure(image=edged)
            panel1.image = image
            panel2.image = edged
            panel3.image = image
            panel4.image = edged
            panel5.image = edged
            
# initialize the window toolkit along with the two image panels
selector = Tk()
panel1 = None
panel2 = None
panel3 = None
panel4 = None
panel5 = None
panel6 = None
panel7 = None
panel8 = None
panel9 = None
# create a button, then when pressed, will trigger a file chooser
# dialog and allow the user to select an input image; then add the
# button the GUI
btn = Button(selector, text="Select an image", command=select_image)
btn.pack(side="bottom", fill="both", expand="yes", padx="10", pady="10")
# kick off the GUI
selector.mainloop()
```
