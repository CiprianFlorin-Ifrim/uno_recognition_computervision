```python
from tkinter import *
from PIL import Image, ImageTk
import cv2
import numpy as np

image_name = './uno_images/b0.jpg'

image = cv2.imread(image_name)

#Rearrang the color channel
b,g,r = cv2.split(image)
img = cv2.merge((r,g,b))

# A root window for displaying objects
root = Tk()  

# Convert the Image object into a TkPhoto object
im = Image.fromarray(img)
imgtk = ImageTk.PhotoImage(image=im) 

# Put it in the display window
Label(root, image=imgtk).pack() 

root.mainloop() # Start the GUI




```

```python
from tkinter import *
class Window(Frame):
    def __init__(self, master=None):
        Frame.__init__(self, master)        
        self.master = master

        # Add image file
        bg = PhotoImage( file = "Computer-Vision_Blog-scaled.jpeg")

        # Show image using label
        label1 = Label( root, image = bg)
        label1.place(x = 0,y = 0)
        # create button, link it to clickExitButton()
        exitButton = Button(self, text="Exit", height = 10, width = 20, command=self.clickExitButton)

        # place button at (0,0)
        exitButton.place(x=30, y=30)

    def clickExitButton(self):
        root.destroy()
        
        
        
root = Tk()
app = Window(root)
root.wm_title("UNO Recognition Software")
root.geometry("200x500")
root.mainloop()
```

```python
# Import module 
from tkinter import *
from IPython.core.display import display, HTML                                    
display(HTML("<style>.container { width:100% !important; }</style>"))                                                #change width of Jupyer Notebook to use the whole window resolution available
  



# Create object 
root = Tk()
  
# Adjust size 
root.geometry("400x400")
  
# Add image file
bg = PhotoImage(file = "Computer-Vision_Blog-scaled.jpeg")
  
# Create Canvas
canvas = Canvas( root, width = 400, height = 400)
  
canvas.pack(fill = "both", expand = True)
  
# Display image
canvas.create_image( 0, 0, image = bg, anchor = "nw")
  
    
frameCnt = 35
frames = [PhotoImage(file='wow.gif',format = 'gif -index %i' %(i)) for i in range(frameCnt)]

def update(ind):
    try:
        frame = frames[ind]
        ind += 1
        label.configure(image=frame)
        root.after(80, update, ind)
    except IndexError:
        pass
    
label = Label(root)
label.pack()
root.after(0, update, 0)
root.mainloop()    
    
# Add Text
canvas.create_text( 200, 250, text = "Welcome")
  
# Create Buttons
button1 = Button( root, text = "Exit")
button3 = Button( root, text = "Start")
button2 = Button( root, text = "Reset")
  
# Display Buttons
button1_canvas = canvas.create_window( 100, 10, anchor = "nw", window = button1)
  
button2_canvas = canvas.create_window( 100, 40, anchor = "nw", window = button2)
  
button3_canvas = canvas.create_window( 100, 70, anchor = "nw", window = button3)
  
# Execute tkinter
root.mainloop()
```

```python
from tkinter import *
from PIL import ImageTk, Image
import cv2

# Create object 
root = Tk()
 
# Adjust size 
root.geometry("1800x480")

# Window Title
root.title('COMPUTER VISION - UNO IDENTIFICATION')

label1 = Label(root)
label1.place(x = 0,y = 0)

# Capture from camera
cap = cv2.VideoCapture(0)

# function for video streaming
def video_stream():
    _, frame = cap.read()
    cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
    img = Image.fromarray(cv2image)
    imgtk = ImageTk.PhotoImage(image=img)
    label1.imgtk = imgtk
    label1.configure(image=imgtk)
    label1.after(1, video_stream) 


video_stream()
root.mainloop()
```
